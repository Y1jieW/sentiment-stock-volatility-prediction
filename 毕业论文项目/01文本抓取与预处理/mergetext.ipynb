```python
import pandas as pd
text_1 = "/Users/wangyijie/Visual_Studio_code/æ¯•ä¸šè®ºæ–‡é¡¹ç›®/01æ–‡æœ¬æŠ“å–ä¸é¢„å¤„ç†/2025_truth.csv"
text_2 = "/Users/wangyijie/Visual_Studio_code/æ¯•ä¸šè®ºæ–‡é¡¹ç›®/01æ–‡æœ¬æŠ“å–ä¸é¢„å¤„ç†/2025_truth2.csv"

df1 = pd.read_csv(text_1)
df2 = pd.read_csv(text_2)

print(df1)
print(df2)
```

```python
from datetime import datetime

# è‡ªå®šä¹‰å‡½æ•°ï¼šè½¬æ¢ "July 23, 2025 @ 11:10 PM ET" ä¸º datetime å¯¹è±¡   
def parse_custom_date(date_str):
    try:
        if pd.isna(date_str):
            return pd.NaT
        clean_str = str(date_str).replace('@', '').replace('ET', '').strip()
        return datetime.strptime(clean_str, '%B %d, %Y %I:%M %p')
    except Exception as e:
        print(f"æ—¥æœŸè§£æå¤±è´¥ï¼š{date_str} -> {e}")
        return pd.NaT



# å‡è®¾æ—¥æœŸåˆ—åä¸º 'date'ï¼Œå¦‚æœä¸åŒè¯·æ›¿æ¢
df1['date_parsed'] = df1['date'].apply(parse_custom_date)
df2['date_parsed'] = df2['date'].apply(parse_custom_date)

# ç­›é€‰ df1 ä¸­ 1æœˆ20æ—¥ä¹‹åçš„æ•°æ®
df1_filtered = df1[df1['date_parsed'] > pd.to_datetime('2025-01-20')]

# ç­›é€‰ df2 ä¸­ 7æœˆ8æ—¥è‡³7æœˆ20æ—¥ä¹‹é—´çš„æ•°æ®ï¼ˆåŒ…å«ï¼‰
start_date = pd.to_datetime('2025-07-08')
end_date = pd.to_datetime('2025-07-20')
df2_filtered = df2[(df2['date_parsed'] >= start_date) & (df2['date_parsed'] <= end_date)]

# åˆå¹¶ä¸¤ä¸ªæ•°æ®é›†å¹¶æŒ‰æ—¥æœŸæ’åº
df_combined = pd.concat([df1_filtered, df2_filtered])
df_combined = df_combined.sort_values(by='date_parsed').reset_index(drop=True)

# è¾“å‡ºç»“æœ
print(df_combined.head())  # æ˜¾ç¤ºå‰å‡ è¡Œæ•°æ®
```
