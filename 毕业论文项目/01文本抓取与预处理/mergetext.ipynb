```python
import pandas as pd
text_1 = "/Users/wangyijie/Visual_Studio_code/毕业论文项目/01文本抓取与预处理/2025_truth.csv"
text_2 = "/Users/wangyijie/Visual_Studio_code/毕业论文项目/01文本抓取与预处理/2025_truth2.csv"

df1 = pd.read_csv(text_1)
df2 = pd.read_csv(text_2)

print(df1)
print(df2)
```

```python
from datetime import datetime

# 自定义函数：转换 "July 23, 2025 @ 11:10 PM ET" 为 datetime 对象   
def parse_custom_date(date_str):
    try:
        if pd.isna(date_str):
            return pd.NaT
        clean_str = str(date_str).replace('@', '').replace('ET', '').strip()
        return datetime.strptime(clean_str, '%B %d, %Y %I:%M %p')
    except Exception as e:
        print(f"日期解析失败：{date_str} -> {e}")
        return pd.NaT



# 假设日期列名为 'date'，如果不同请替换
df1['date_parsed'] = df1['date'].apply(parse_custom_date)
df2['date_parsed'] = df2['date'].apply(parse_custom_date)

# 筛选 df1 中 1月20日之后的数据
df1_filtered = df1[df1['date_parsed'] > pd.to_datetime('2025-01-20')]

# 筛选 df2 中 7月8日至7月20日之间的数据（包含）
start_date = pd.to_datetime('2025-07-08')
end_date = pd.to_datetime('2025-07-20')
df2_filtered = df2[(df2['date_parsed'] >= start_date) & (df2['date_parsed'] <= end_date)]

# 合并两个数据集并按日期排序
df_combined = pd.concat([df1_filtered, df2_filtered])
df_combined = df_combined.sort_values(by='date_parsed').reset_index(drop=True)

# 输出结果
print(df_combined.head())  # 显示前几行数据
```
